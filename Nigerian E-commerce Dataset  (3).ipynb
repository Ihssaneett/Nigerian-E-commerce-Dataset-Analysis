 Order ID is shown as an int64, it's not used for math, it’s just a unique identifier, like a name tag. So it makes more sense to treat it as a string (object)

```python
df['Order ID'] = df['Order ID'].astype(str)
df.info()
```

Now I change the date format of the 'Order Date' column so python recognizes it as a date

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
```

Time to visualize:

```python
daily_sales = df.groupby('Order Date')['Total Price'].sum()
daily_sales.plot(kind='line', figsize=(12, 6), title='Total Sales Over Time')
```

```python
top_items = df.groupby('Item Name')['Total Price'].sum().sort_values(ascending=False).head(10)
print(top_items)
top_items.plot(kind= 'bar', figsize=(12, 6), title='Top 10 Products Quantity Sold')
```

```python
branch_sales = df.groupby('Branch Name')['Total Price'].sum().sort_values(ascending=False)
print(branch_sales)
```

```python
branch_sales.plot(kind='bar', title='Sales by Branch', ylabel='Total Revenue', figsize=(10,6))
```

```python
pivot_branch_sales = pd.pivot_table(df, index='Branch Location', values='Total Price', aggfunc='sum')
print(pivot_branch_sales)
```

```python
pivot_branch_sales.plot(kind='line', figsize=(10,5), title='Total Sales by Branch')
```

```python
pivot_avg_price = pd.pivot_table(df, index='Branch Location', values='Item Price', aggfunc='mean')
print(pivot_avg_price)
```

```python
pivot_avg_price.plot(kind='bar', title='Average Item Price per Branch', ylabel='Average Price', figsize=(6, 5), legend=False)
```

```python
df['Order Month'] = df['Order Date'].dt.to_period('M')
pivot_monthly_sales = pd.pivot_table(df, index='Order Month', values='Total Price', aggfunc='sum')
print(pivot_monthly_sales)
```

```python
pivot_monthly_sales.plot(kind='line', figsize=(10,5), title='Monthly Sales Trend')
```

```python
pivot_status_sales = pd.pivot_table(df, index='Item Status', values='Total Price', aggfunc='sum')
print(pivot_status_sales)
```

```python
pivot_status_sales.plot(kind='bar', title='Sales by Item Status', ylabel='Total Revenue', figsize=(6, 5), legend=False)
```

```python
status_counts = df["Item Status"].value_counts()
plt.figure(figsize=(6,6))
status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=["#4CAF50", "#0000FF", "#FFFF00","#FF5722" ])
plt.title("Order Status Distribution")
plt.ylabel('')
plt.tight_layout()
plt.show()
```
